```python
# Pacotes necessários.

import matplotlib.pyplot as plt
import numpy as np
import pygame
import random
from pygame.locals import *
```

```python
class CrocoBoy():
    def __init__(self) -> None:
        self.color = (30, 200, 20)
        self.vel = np.array([10, -10])*5
        self.s0 = np.array([50,200])
        
    def reset(self):
        objs, self.vel = self.s0, 0
```

```python
pygame.init()
from functions import *

# Tamanho da tela e definição do FPS
screen = pygame.display.set_mode((1280, 720))
clock = pygame.time.Clock()
FPS = 60 

load_assets()

BLACK = (0, 0, 0)

personagem = CrocoBoy()

# Inicializar posicoes

objs = personagem.s0

objv = np.array([10, -10])*10
k = 1000

a = np.array([0, 0.2])

planeta = np.array([300,200]) 
lua = np.array([300,250])

# Personagem
personagem = pygame.Surface((5, 5))  # Tamanho do personagem
personagem.fill(personagem.color)  # Cor do personagem

rodando = True
while rodando:
    # Capturar eventos
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            rodando = False

    y = pygame.mouse.get_pos()

    if objs[0] < 10 or objs[0]>390 or objs[1]<10 or objs[1]>390: # Se eu chegar ao limite da tela, reinicio a posição do personagem
        personagem.reset()
        

    # Controlar frame rate
    clock.tick(FPS)

    # Processar posicoes
    for i in range(0,50):
        dp = planeta - objs
        d = np.linalg.norm(dp) # moduulo

        a = (dp/d)*k/d**2

        dp = lua - objs
        d = np.linalg.norm(dp) # moduulo

        a2 = (dp/d)*(k/2)/d**2

        objv[i] = objv[i] + a + a2
        objs = objs + 0.1 * objv[i] # s = s(-1) + a

    # Desenhar fundo
    screen.fill(BLACK)

    # Desenhar personagem
    for i in range(0,50):
        rect = pygame.Rect(objs, (10, 10))  # First tuple is position, second is size.
        screen.blit(personagem, rect)
    circle = pygame.draw.circle(surface = screen, radius = 25,center=planeta,color='yellow')
    circle2 = pygame.draw.circle(surface = screen, radius = 10,center=lua,color='blue')
    # Update!
    pygame.display.update()

# Terminar tela
pygame.quit()
```

```python
pygame.init()

# Tamanho da tela e definição do FPS
screen = pygame.display.set_mode((1280, 720))
clock = pygame.time.Clock()
FPS = 60  # Frames per Second

BLACK = (0, 0, 0)
COR_PERSONAGEM = (30, 200, 20)

# Inicializar posicoes
s0 = np.array([50,200])
v0 = np.array([10, -10])*0.5
a = np.array([0, 0.4])
v = [v0 for i in range(50)]
s = [s0+5*np.random.randn(2) for i in range(50)]
planeta = np.array([200, 200])


# Personagem
personagem = pygame.Surface((5, 5))  # Tamanho do personagem
personagem.fill(COR_PERSONAGEM)  # Cor do personagem

rodando = True
while rodando:
    # Capturar eventos
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            rodando = False

    for i in range(50):
        if s[i][0]<10 or s[i][0]>390 or s[i][1]<10 or s[i][1]>390: # Se eu chegar ao limite da tela, reinicio a posição do personagem
            s[i] = s0
            pos_mouse = pygame.mouse.get_pos()
            y = pos_mouse - s0
            y = y / np.linalg.norm(y)
            y = y * 10
            rnd = np.random.randn(2)
            escala = 10
            v[i] = y + rnd * escala


    # Controlar frame rate
    clock.tick(FPS)

    # Processar posicoes
    for i in range(50):
        v[i] = v[i] + a
        s[i] = s[i] + 0.1 * v[i]

    # Desenhar fundo
    screen.fill(BLACK)

    # Desenhar personagem
    for i in range(50):
        C = 1000
        direcao_a = planeta - s[i]
        d = np.linalg.norm(direcao_a)
        direcao_a = direcao_a / d

        mag_a = C / d**2
        a = direcao_a * mag_a

        rect = pygame.Rect(s[i], (10, 10))  # First tuple is position, second is size.
        screen.blit(personagem, rect)

    # Update!
    pygame.display.update()

# Terminar tela
pygame.quit()
```
